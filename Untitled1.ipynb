a) For a single element in C(i,j), we perform k multiplications (for each corresponding element in the row of A and the column of B) and then sum up these k products with k-1 additions. So, in total, we need k multiplications and k-1 additions for one element of C.

b) C is an (n, m) matrix,we have n*m elements to compute.Therefore, the total number of multiplications needed is n*m*k, and the total number of additions needed is n*m*(k-1).
 
  hence we can say that,for the matrix multiplication of an (n,k) matrix and a (k,m) matrix,we need n*m*k multiplications and n*m*(k-1) additions.

Question 3

Finding the highest element in a list requires one pass of the array. Finding the second highest element requires 2 passes of the the array. Time complexity for finding median would be O(n^2).where n is the size of array.

And the sorting algorithm will be more efficient for this and the time complexity for this will be O(nlogn).


```python
import numpy as np
import time
from typing import List

def find_median(arr:List[int])->float:
    n=len(arr)
    for i in range(n//2):
        max_idx=i
        for j in range(i+1,n):
            if arr[j]>arr[max_idx]:
                max_idx=j
        arr[i],arr[max_idx]=arr[max_idx], arr[i]
    if n%2==0:
        return (arr[n//2-1]+arr[n//2])/2
    else:
        return arr[n//2]

def find_median_sort(arr: List[int])->float:
    arr.sort()
    n=len(arr)
    if n%2==0:
        return (arr[n//2-1]+arr[n//2])/2
    else:
        return arr[n // 2]

def find_median_np(arr: List[int]) -> float:
    return np.median(arr)

# Test the functions
for i in range(1,7):
    n=10**i
    arr=list(range(n))
    start_time=time.time()
    find_median(arr)
    print(f"Method 1: n={n},time taken={time.time()-start_time:.6f} seconds")

    start_time = time.time()
    find_median_sort(arr)
    print(f"Method 2:n={n},time taken={time.time()-start_time:.6f} seconds")

    arr_np=np.array(arr)
    start_time=time.time()
    find_median_np(arr_np)
    print(f"Method 3: n={n},time taken={time.time()-start_time:.6f} seconds")
    print()
```

```python
#Question 7
students_data = {}

def data(year,branch,roll,name,marks):
    if year not in students_data:
        students_data[year]={}
    
    if branch not in students_data[year]:
        students_data[year][branch]={}
    
    students_data[year][branch][roll]={
        'Roll Number':roll,
        'Name':name,
        'Marks':marks
    }

def display():
    for year, branches in students_data.items():
        print(f"{year}:")
        for branch, students in branches.items():
            print(f"  {branch}:")
            for roll, student_info in students.items():
                print(f"        Roll Number: {student_info['Roll Number']}, Name: {student_info['Name']}, Marks:")
                for subject, mark in student_info['Marks'].items():
                    print(f"            {subject}: {mark}")
                print()

def main():
    while True:
        year = int(input("Enter the academic year (e.g., 2022): "))
        branch = input("Enter the branch (e.g., Branch 1): ")
        roll= int(input("Enter the roll number: "))
        name = input("Enter the student's name: ")
        math_marks = int(input("Enter Math marks: "))
        english_marks = int(input("Enter English marks: "))

        marks = {'Maths': math_marks, 'English': english_marks}
        data(year, branch, roll, name, marks)

        print("\nUpdated Student Data:")
        display()

        continue_input = input("Do you want to continue? (yes/no): ").lower()
        if continue_input != 'yes':
            break

if __name__ == "__main__":
    main()
```

```python
#Question 8
class Student:
    def __init__(self,roll,name,marks):
        self.roll=roll
        self.name=name
        self.marks=marks

class Branch:
    def __init__(self,name):
        self.name=name
        self.students=[]

    def add_student(self,student):
        self.students.append(student)

class Year:
    def __init__(self,year):
        self.year=year
        self.branches=[]

    def add_branch(self,branch):
        self.branches.append(branch)
        
def display(years):
    for year in years:
        print(f"{year.year}:")
        for branch in year.branches:
            print(f"  {branch.name}:")
            for student in branch.students:
                print(f"        Roll Number: {student.roll}, Name: {student.name}, Marks:")
                for subject, mark in student.marks.items():
                    print(f"     {subject}: {mark}")
                print()

def main():
    years=[]

    while True:
        year_input=int(input("Enter the academic year (e.g., 2022): "))
        year=Year(year_input)

        branch_input=input("Enter the branch (e.g., Branch 1): ")
        branch=Branch(branch_input)

        roll_input=int(input("Enter the roll number: "))
        name_input=input("Enter the student's name: ")
        math_marks_input=int(input("Enter Math marks: "))
        english_marks_input=int(input("Enter English marks: "))

        marks_input={'Maths': math_marks_input, 'English': english_marks_input}

        student=Student(roll_input, name_input, marks_input)
        branch.add_student(student)
        year.add_branch(branch)

        years.append(year)

        print("\nUpdated Student Data:")
        display(years)

        continue_input = input("Do you want to continue? (yes/no): ").lower()
        if continue_input!='yes':
            break

if __name__=="__main__":
    main()
```

```python
#question 9
import matplotlib.pyplot as plt
import numpy as np

x = np.arange(0.5, 100.1, 0.5)

y1 = x
y2 = x**2
y3 = (x**3)/100 
y4 = np.sin(x)
y5 = np.sin(x)/x
y6 = np.log(x)
y7 = np.exp(x)

plt.figure(figsize=(10, 15))

plt.subplot(4, 2, 1)
plt.plot(x, y1)
plt.title('y = x')

plt.subplot(4, 2, 2)
plt.plot(x, y2)
plt.title('y = x^2')

plt.subplot(4, 2, 3)
plt.plot(x, y3) 
plt.title('y = (x^3)/100')

plt.subplot(4, 2, 4)
plt.plot(x, y4)
plt.title('y = sin(x)')

plt.subplot(4, 2, 5)
plt.plot(x, y5)
plt.title('y = sin(x)/x')

plt.subplot(4, 2, 6)
plt.plot(x, y6)
plt.title('y = log(x)')

plt.subplot(4, 2, 7)
plt.plot(x, y7)
plt.title('y = e^x')

plt.tight_layout()
plt.show()
```

```python
#question 11
import pandas as pd
import numpy as np

data = {
    'a': [1, 2, 3, 4, 5],
    'b': [2, 3, 4, 5, 6],
    'c': [3, 4, 5, 6, 7],
    'd': [4, 5, 6, 7, 8],
    'e': [5, 6, 7, 8, 9]
}
df=pd.DataFrame(data)

df['f']=df[['a', 'b', 'c', 'd', 'e']].sum(axis=1)

df['g']=np.where(df['f'] < 8, 'LT8', 'GT8')

lt8_count=df[df['g']=='LT8'].shape[0]

std_lt8=df[df['g']=='LT8']['f'].std()
std_gt8=df[df['g']=='GT8']['f'].std()

print("Number of rows where 'g' is 'LT8':",lt8_count)
print("Standard deviation of 'f' where 'g' is 'LT8':",std_lt8)
print("Standard deviation of 'f' where 'g' is 'GT8':",std_gt8)
```

```python
#question 13
import numpy as np

def custom_argmin(arr):
    min_index=0
    min_value=arr[0]

    for i in range(1, len(arr)):
        if arr[i]<min_value:
            min_value=arr[i]
            min_index=i

    return min_index

arr=np.array([3, 1, 4, 2, 5])
index=custom_argmin(arr)
print("Custom argmin index:",index)

np_index=np.argmin(arr)
print("np.argmin index:",np_index)
```
